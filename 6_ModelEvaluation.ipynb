<h2> Splitting Data into Train and Test set

```python
import pandas as pd
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df = pd.read_csv('clean_df.csv')
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test = tts(df[['highway-mpg']],df['price'],test_size = 0.3)
print(x_train)
print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(lr,df[['highway-mpg']],df['price'],cv=3)
score
```

```python
avg = score.mean()
avg
```

```python
from sklearn.model_selection import cross_val_predict
prdc = cross_val_predict(lr,df[['highway-mpg']],df['price'],cv=3)
print(prdc)
```
